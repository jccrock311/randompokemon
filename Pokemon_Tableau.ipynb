This small project is to demonstrate my abilities to import restful APIs, sort existing lists, make new lists, utilize new lists (through random number generation) and jsonify the new lists to import into Tableau for research purposes.

```python
import requests
import random

response = requests.get('https://pokeapi.co/api/v2/pokemon-species?limit=100000&offset=0')
print(response)
```

```python
print(response.json())
```

```python
# create seperate ins for specific criteria used

pokemon = response.json()['results']
pokemon_list = []

for x in pokemon:
    pokemon_list.append(x)
    print(pokemon_list)
    

# my config limit was pre-set, as indicated by the warning below, but the data is there.
```

```python
print(pokemon_list[0])
print(pokemon_list[472]['name'])
print(pokemon_list[1003]['url'])
```

```python
# create 20 random instances from the full list of Pokemon and then print them

random_pokemon = random.choices(pokemon_list, k=20)
print("The 20 random Pokemon are: " +  str(random_pokemon))
```
